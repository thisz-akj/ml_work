```python
from sklearn.datasets import fetch_openml
```

```python
from sklearn.datasets import fetch_california_housing
housing=fetch_california_housing()
```

```python
housing
```

```python
pd.DataFrame(housing.data)
```

```python
dataset=pd.DataFrame(housing.data)
```

```python
dataset
```

```python
housing.feature_names
```

```python
dataset.columns=housing.feature_names
```

```python
dataset
```

```python
#independent features and dependent features
x=dataset#input->dataframe of housing
y=housing.target#required output
```

```python
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.30, random_state=42)
```

```python
X_train
```

```python
#standardizing the dataset for better output
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)#fitting old values and updatting it in training sample
```

```python
X_test=scaler.transform(X_test)
#why fit is not used here?
#The point of having a test set is to have an accurate measure of how well the model will do on unseen data. 
#If you include the test set in the training data,
#then the test accuracy will be artificially inflated
```

```python
#to get older data
scaler.inverse_transform(X_train)
```

```python
#applying linear regression
from sklearn.linear_model import LinearRegression
#cross validation
from sklearn.model_selection import cross_val_score
```

```python
#creating model
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
#crossvalidating to check accuracy
cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=5)
```

```python
#this is mean squared error/ you can say we are checking accuracy for different cases i.e cv
mse=cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=5)
#taking mean of all those op for n cv
np.mean(mse)
#output should be less for good
```

```python
#prediction of test data from model
reg_predict=regression.predict(X_test)
```

```python
reg_predict #for each tesst data it will give output
```
